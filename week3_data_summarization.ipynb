## Part 2: Data Summarization

```python
# Total "Gross" of all top 1000 movies

imdb["Gross"].sum()
```

```python
# Average "No_of_Votes" of all movies

imdb["No_of_Votes"].mean()
```

```python
# Count movies in each decade (e.g., ..., 1980, 1990, 2000, 2010, 2020)
# Sort decades by the number of movies in descending order

imdb["Decade"].value_counts()
```

```python
# Count movies by different directors.
# Show the top 10 directors with the most movies in this list

imdb["Director"].value_counts().head(10)
```

```python
# For each director, calculate the number of movies, average IMDB_Rating, and total gross.
# Ranked in descending order of total gross
# Show the top 10 directors with the most gross

imdb.groupby("Director").agg({
    "Series_Title": "count",
    "IMDB_Rating": "mean",
    "Gross": "sum"
}).rename(columns={
    "Series_Title": "Movie_Count",
    "IMDB_Rating": "Average_IMDB_Rating",
    "Gross": "Total_Gross"
}).sort_values("Total_Gross", ascending=False).head(10)
```

```python
# Group movies by decade and director
# In each group (i.e., for each decade and each director),
# calculate the number of movies and average IMDB rating.
# Sort in descending order of movie count

imdb.groupby(["Decade", "Director"]).agg({
    "Series_Title": "count",
    "IMDB_Rating": "mean"
}).rename(columns={
    "Series_Title": "Movie_Count",
    "IMDB_Rating": "Average_IMDB_Rating"
}).sort_values("Movie_Count", ascending=False)
```

```python
# Find the top 3 highest rated movies of each year since 2010.
# Show their released year, ranking, title, and IMDB rating.
# Sort them in descending order of year and ascending order of ranking

imdb_since_2010 = imdb[imdb["Released_Year"] >= 2010].copy()
imdb_since_2010["Ranking"] = imdb_since_2010.groupby("Released_Year")["IMDB_Rating"].rank(ascending=False, method="first").astype(int)
imdb_since_2010[imdb_since_2010["Ranking"] <= 3][["Released_Year", "Ranking", "Series_Title", "IMDB_Rating"]].sort_values(["Released_Year", "Ranking"], ascending=[False, True])
```

```python
# Find all directors whose movies appeared in at least five different decades.
# Your result should include: director, decade, and the number of movies in the decade.

director_decade = imdb.groupby(["Director", "Decade"]).size().reset_index(name="Movie_Count")
five_decades = director_decade.groupby("Director")["Decade"].nunique()
director_decade[director_decade["Director"].isin(five_decades[five_decades >= 5].index)].sort_values(["Director", "Decade"])
```
